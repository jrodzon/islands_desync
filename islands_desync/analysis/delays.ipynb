```python
import glob

files = glob.glob('*Imigrants.json')
files.sort()
files
```

```python
import json
import pandas

data = {}

for result_file in files:
    with open(result_file) as f:
        result = json.load(f)

        for step, value in result.items():
            data_at_step = data.get(step, [])
            data_at_step.extend(value["iteration_numbers"])
            data[step] = data_at_step


print(data)
df = pandas.DataFrame.from_dict(data, orient='index')
print(df)
df.plot()

```
